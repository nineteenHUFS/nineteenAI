```python
import pandas as pd
import sqlite3
import re
from google.colab import files

# ÌÖçÏä§Ìä∏ ÌååÏùº Í≤ΩÎ°ú ÏÑ§Ï†ï
txt_file_path = "tc.txt"

# ÌÖçÏä§Ìä∏ ÌååÏùº ÏùΩÍ∏∞
with open(txt_file_path, 'r', encoding='utf-8') as file:
    data = file.read()

# Ï†ïÍ∑ú ÌëúÌòÑÏãùÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú
pattern = re.compile(
    r'product Name: (?P<name>.*?)\n'
    r'Post \d+ Uploader: (?P<uploader>.*?)\n'
    r'Post \d+ Follower: (?P<follower>.*?)\n'
    r'Post \d+ Title: (?P<title>.*?)\n'
    r'Post \d+ URL: (?P<url>.*?)\n'
    r'Post \d+ Hashtags: (?P<hashtag>.*?)\n'
    r'Post \d+ Like: (?P<like>\d+)\n'
    r'Post \d+ Comments: (?P<comments>\d+)\n'
    r'Post \d+ Bookmark: (?P<bookmark>\d+)\n'
    r'Post \d+ Share: (?P<share>\d+)\n'
    r'Post \d+ Videocount: (?P<videocount>.*?)\n'
)

matches = pattern.finditer(data)

# Îç∞Ïù¥ÌÑ∞Î•º Ï†ÄÏû•Ìï† Î¶¨Ïä§Ìä∏ ÏÉùÏÑ±
records = []

for match in matches:
    record = {
        'name': match.group('name').strip(),
        'influencer': match.group('uploader').strip(),
        'follower': match.group('follower').strip(),
        'title': match.group('title').strip(),
        'url': match.group('url').strip(),
        'hashtag': match.group('hashtag').strip(),
        'like': int(match.group('like')),
        'comments': int(match.group('comments')),
        'bookmark': int(match.group('bookmark')),
        'share': int(match.group('share')),
        'videocount': match.group('videocount').strip(),
    }
    records.append(record)

# Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
df = pd.DataFrame(records)

# ÌåîÎ°úÏõå ÏàòÏôÄ ÎπÑÎîîÏò§ ÏàòÏùò Îã®ÏúÑ Î≥ÄÌôò (Ïòà: K, M Îã®ÏúÑ Î≥ÄÌôò)
def convert_to_number(value):
    value = value.upper()
    if 'K' in value:
        return float(value.replace('K', '')) * 1000
    elif 'M' in value:
        return float(value.replace('M', '')) * 1000000
    try:
        return float(value)
    except ValueError:
        return value

df['follower'] = df['follower'].apply(convert_to_number)
df['videocount'] = df['videocount'].apply(convert_to_number)

# SQLite Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ ÌååÏùºÎ°ú Ï†ÄÏû•
conn = sqlite3.connect("influencer_data.db")  # Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ ÌååÏùº ÏÉùÏÑ± (influencer_data.db)
df.to_sql('influencer_data', conn, if_exists='replace', index=False)  # 'influencer_data' ÌÖåÏù¥Î∏îÎ°ú Ï†ÄÏû•
conn.close()  # Ïª§ÎÑ•ÏÖò Ï¢ÖÎ£å

# Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ ÌååÏùº Îã§Ïö¥Î°úÎìú
files.download('influencer_data.db')  # ÌååÏùº Îã§Ïö¥Î°úÎìú

print("ÌÖçÏä§Ìä∏ ÌååÏùºÏù¥ ÏÑ±Í≥µÏ†ÅÏúºÎ°ú Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§Î°ú Î≥ÄÌôòÎêòÏóàÍ≥†, influencer_data.db ÌååÏùºÏù¥ Îã§Ïö¥Î°úÎìúÎê©ÎãàÎã§.")
```

```python
import sqlite3
import pandas as pd

# ÌÜµÌï©Ìï† Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ ÌååÏùºÎì§Ïùò Í≤ΩÎ°úÎ•º Î¶¨Ïä§Ìä∏Î°ú Ï†ïÏùò
db_files = [
    "2page_products.db",
    "3page_products.db",
    "4page_products.db",
    "5page_products.db",
    "6page_products.db",
    "7page_products.db",
    "8page_products.db",
    "9page_products.db"
]

# ÌÜµÌï©Îêú Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ ÌååÏùº Í≤ΩÎ°ú ÏÑ§Ï†ï
OUTPUT_DB = "all_merged_database.db"  # ÏµúÏ¢Ö ÌÜµÌï©Îêú Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ ÌååÏùºÎ™Ö

# ÌÜµÌï©Ìï† ÌÖåÏù¥Î∏î Ïù¥Î¶Ñ ÏÑ§Ï†ï
TABLE_NAME = "products"

# Ï∂úÎ†• Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ Ï¥àÍ∏∞Ìôî Î∞è ÌÖåÏù¥Î∏î ÏÉùÏÑ±
def initialize_output_database():
    conn = sqlite3.connect(OUTPUT_DB)
    cursor = conn.cursor()

    # ÌÖåÏù¥Î∏î Ï¥àÍ∏∞Ìôî
    cursor.execute(f"DROP TABLE IF EXISTS {TABLE_NAME}")

    # Í∏∞Ï°¥ Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§Ïùò ÌÖåÏù¥Î∏î Íµ¨Ï°∞ÏôÄ ÎèôÏùºÌïòÍ≤å ÏÉà ÌÖåÏù¥Î∏î ÏÉùÏÑ±
    cursor.execute(f'''
        CREATE TABLE IF NOT EXISTS {TABLE_NAME} (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            name TEXT,
            sales TEXT,
            rating REAL,
            price REAL
        )
    ''')

    conn.commit()
    conn.close()
    print(f"Output database '{OUTPUT_DB}' initialized successfully.")

# Í∞Å Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ ÌååÏùºÏóêÏÑú Îç∞Ïù¥ÌÑ∞ ÏùΩÍ≥† ÌÜµÌï©
def merge_databases():
    all_data = []

    for db_file in db_files:
        print(f"Processing file: {db_file}")
        conn = sqlite3.connect(db_file)
        query = f"SELECT * FROM {TABLE_NAME}"
        try:
            data = pd.read_sql_query(query, conn)
            all_data.append(data)
            print(f"Successfully read data from {db_file}")
        except Exception as e:
            print(f"Failed to read from {db_file}. Error: {e}")
        finally:
            conn.close()

    if all_data:
        # Îç∞Ïù¥ÌÑ∞ ÌÜµÌï©
        merged_data = pd.concat(all_data, ignore_index=True)
        print("Data from all databases merged successfully.")

        # Í∏∞Ï°¥ Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ÏóêÏÑú 'id' Ïª¨ÎüºÏùÄ ÏûêÎèô Ï¶ùÍ∞ÄÎêòÎØÄÎ°ú Ï†úÍ±∞ ÌõÑ Ï†ÄÏû•
        if 'id' in merged_data.columns:
            merged_data = merged_data.drop(columns=['id'])

        # ÌÜµÌï©Îêú Îç∞Ïù¥ÌÑ∞Î•º ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§Ïóê Ï†ÄÏû•
        save_to_output_database(merged_data)

# ÌÜµÌï©Îêú Îç∞Ïù¥ÌÑ∞Î•º Ï∂úÎ†• Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§Ïóê Ï†ÄÏû•
def save_to_output_database(merged_data):
    conn = sqlite3.connect(OUTPUT_DB)
    merged_data.to_sql(TABLE_NAME, conn, if_exists='append', index=False)
    conn.close()
    print(f"Data saved to output database '{OUTPUT_DB}' successfully.")

# Î©îÏù∏ Ìï®Ïàò
if __name__ == "__main__":
    initialize_output_database()
    merge_databases()
```

```python
import pandas as pd
import sqlite3
import re

# ÌÖçÏä§Ìä∏ ÌååÏùº Í≤ΩÎ°ú ÏÑ§Ï†ï
txt_file_path = "4page_data.txt"

# ÌÖçÏä§Ìä∏ ÌååÏùº ÏùΩÍ∏∞
with open(txt_file_path, 'r', encoding='utf-8') as file:
    data = file.read()

# Ï†ïÍ∑ú ÌëúÌòÑÏãùÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú
pattern_product = re.compile(r'product Name: (.+)')
pattern_post = re.compile(
    r'Post \d+ Uploader: (?P<uploader>.+)\n'
    r'Post \d+ Follower: (?P<follower>.+)\n'
    r'Post \d+ Title: (?P<title>.+)\n'
    r'Post \d+ URL: (?P<url>.+)\n'
    r'Post \d+ Hashtags: (?P<hashtag>.+)\n'
    r'Post \d+ Like: (?P<like>\d+)\n'
    r'Post \d+ Comments: (?P<comments>\d+)\n'
    r'Post \d+ Bookmark: (?P<bookmark>\d+)\n'
    r'Post \d+ Share: (?P<share>\d+)\n'
    r'Post \d+ Videocount: (?P<videocount>.+)\n'
)

# Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú
records = []
current_product_name = None

lines = data.splitlines()
for i, line in enumerate(lines):
    product_match = pattern_product.match(line)
    if product_match:
        current_product_name = product_match.group(1).strip()
        continue

    post_match = pattern_post.match("\n".join(lines[i:i+11]))
    if post_match and current_product_name:
        record = {
            'name': current_product_name,
            'influencer': post_match.group('uploader').strip(),
            'follower': post_match.group('follower').strip(),
            'title': post_match.group('title').strip(),
            'url': post_match.group('url').strip(),
            'hashtag': post_match.group('hashtag').strip(),
            'like': int(post_match.group('like')),
            'comments': int(post_match.group('comments')),
            'bookmark': int(post_match.group('bookmark')),
            'share': int(post_match.group('share')),
            'videocount': post_match.group('videocount').strip(),
        }
        records.append(record)

# Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
df = pd.DataFrame(records)

# Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏù¥ ÎπÑÏñ¥ ÏûàÎäîÏßÄ ÌôïÏù∏
if df.empty:
    print("Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏù¥ ÎπÑÏñ¥ ÏûàÏäµÎãàÎã§. Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú Í≥ºÏ†ïÏùÑ ÌôïÏù∏ÌïòÏÑ∏Ïöî.")
else:
    # Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Ïó¥ Î™©Î°ùÍ≥º ÏùºÎ∂Ä Îç∞Ïù¥ÌÑ∞ ÏÉòÌîå ÌôïÏù∏
    print("Ï∂îÏ∂úÎêú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùò Ïó¥ Î™©Î°ù:", df.columns)
    print("Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏùºÎ∂Ä ÏÉòÌîå:\n", df.head())

    # ÌåîÎ°úÏõå ÏàòÏôÄ ÎπÑÎîîÏò§ ÏàòÏùò Îã®ÏúÑ Î≥ÄÌôò (Ïòà: K, M Îã®ÏúÑ Î≥ÄÌôò)
    def convert_to_number(value):
        value = value.upper()
        if 'K' in value:
            return float(value.replace('K', '')) * 1000
        elif 'M' in value:
            return float(value.replace('M', '')) * 1000000
        try:
            return float(value)
        except ValueError:
            return value

    # ÌåîÎ°úÏõåÏôÄ ÎπÑÎîîÏò§ ÏàòÏùò ÌòïÏãù Î≥ÄÌôò ÏãúÎèÑ
    if 'follower' in df.columns:
        df['follower'] = df['follower'].apply(convert_to_number)
    if 'videocount' in df.columns:
        df['videocount'] = df['videocount'].apply(convert_to_number)

    # SQLite Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ ÌååÏùºÎ°ú Ï†ÄÏû•
    db_path = "tiktok_influencer_data2.db"
    conn = sqlite3.connect(db_path)  # Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ ÌååÏùº ÏÉùÏÑ±
    df.to_sql('influencer_data', conn, if_exists='replace', index=False)  # 'influencer_data' ÌÖåÏù¥Î∏îÎ°ú Ï†ÄÏû•
    conn.close()

    print(f"Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§Í∞Ä '{db_path}' Í≤ΩÎ°úÏóê ÏÑ±Í≥µÏ†ÅÏúºÎ°ú Ï†ÄÏû•ÎêòÏóàÏäµÎãàÎã§.")
```

```python
import sqlite3
import pandas as pd

# Î≥ëÌï©Îêú Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ ÌååÏùº Í≤ΩÎ°ú ÏÑ§Ï†ï
merged_db_path = "merged_database.db"  # Î°úÏª¨Ïóê ÏûàÎäî ÌååÏùº Í≤ΩÎ°úÎ°ú ÏÑ§Ï†ïÌïòÏÑ∏Ïöî

# Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§Ïóê ÏûàÎäî ÌÖåÏù¥Î∏î Î™©Î°ù ÌôïÏù∏ Ìï®Ïàò Ï†ïÏùò
def list_tables(db_path):
    conn = sqlite3.connect(db_path)
    tables = conn.execute("SELECT name FROM sqlite_master WHERE type='table';").fetchall()
    conn.close()
    return [table[0] for table in tables]

# Î≥ëÌï©Îêú Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§Ïùò ÌÖåÏù¥Î∏î Î™©Î°ù Ï∂úÎ†•
merged_tables = list_tables(merged_db_path)
print("merged_data.db ÌÖåÏù¥Î∏î Î™©Î°ù:", merged_tables)

# Í∞Å ÌÖåÏù¥Î∏îÏùò ÎÇ¥Ïö©ÏùÑ Ï∂úÎ†•ÌïòÎäî Ìï®Ïàò Ï†ïÏùò
def display_table_content(db_path, table_name):
    conn = sqlite3.connect(db_path)
    query = f"SELECT * FROM {table_name} LIMIT 10"
    df = pd.read_sql(query, conn)
    conn.close()
    return df

# Î≥ëÌï©Îêú Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ ÌÖåÏù¥Î∏î ÎÇ¥Ïö© Ï∂úÎ†•
for table in merged_tables:
    print(f"merged_data.db - ÌÖåÏù¥Î∏î '{table}' ÎÇ¥Ïö©:")
    merged_content = display_table_content(merged_db_path, table)
    print(merged_content)
```

```python
import sqlite3
import pandas as pd

# Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ ÌååÏùº Í≤ΩÎ°ú ÏÑ§Ï†ï
DB_FILE_1 = "tiktok_influencer_data2.db"  # Ï≤´ Î≤àÏß∏ Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ ÌååÏùº Í≤ΩÎ°ú
DB_FILE_2 = "all_merged_database.db"  # Îëê Î≤àÏß∏ Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ ÌååÏùº Í≤ΩÎ°ú

# ÌÖåÏù¥Î∏î Ïù¥Î¶Ñ ÏÑ§Ï†ï
TABLE_NAME_1 = "influencer_data"  # Ï≤´ Î≤àÏß∏ Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ÏóêÏÑúÏùò ÌÖåÏù¥Î∏î Ïù¥Î¶Ñ
TABLE_NAME_2 = "products"  # Îëê Î≤àÏß∏ Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ÏóêÏÑúÏùò ÌÖåÏù¥Î∏î Ïù¥Î¶Ñ

# Îëê ÌÖåÏù¥Î∏îÏùÑ Î∂àÎü¨ÏôÄ 'name' Ïª¨ÎüºÏùÑ Í∏∞Ï§ÄÏúºÎ°ú Î≥ëÌï©
def merge_tables_from_dbs():
    # Ï≤´ Î≤àÏß∏ Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ Ïó∞Í≤∞ Î∞è Îç∞Ïù¥ÌÑ∞ Î°úÎìú
    conn1 = sqlite3.connect(DB_FILE_1)
    query1 = f"SELECT * FROM {TABLE_NAME_1}"
    df1 = pd.read_sql_query(query1, conn1)
    conn1.close()
    print(f"Data from {TABLE_NAME_1} loaded successfully.")

    # Îëê Î≤àÏß∏ Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ Ïó∞Í≤∞ Î∞è Îç∞Ïù¥ÌÑ∞ Î°úÎìú
    conn2 = sqlite3.connect(DB_FILE_2)
    query2 = f"SELECT * FROM {TABLE_NAME_2}"
    df2 = pd.read_sql_query(query2, conn2)
    conn2.close()
    print(f"Data from {TABLE_NAME_2} loaded successfully.")

    # 'name' Ïª¨ÎüºÏùÑ Í∏∞Ï§ÄÏúºÎ°ú Î≥ëÌï© (inner join)
    merged_df = pd.merge(df1, df2, on='name', how='inner')

    # Î≥ëÌï©Îêú Îç∞Ïù¥ÌÑ∞ Ï∂úÎ†•
    print("Merged DataFrame:")
    print(merged_df)

    return merged_df

# Î≥ëÌï©Îêú Îç∞Ïù¥ÌÑ∞Î•º ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§Ïóê Ï†ÄÏû•
OUTPUT_DB = "merged_database.db"  # ÏµúÏ¢Ö ÌÜµÌï©Îêú Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ ÌååÏùºÎ™Ö
MERGED_TABLE_NAME = "merged_table"

def save_merged_data(merged_data):
    conn = sqlite3.connect(OUTPUT_DB)
    cursor = conn.cursor()

    # Î≥ëÌï©Îêú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ Ïù¥Ïö©ÌïòÏó¨ ÎèôÏ†ÅÏúºÎ°ú CREATE TABLE ÏøºÎ¶¨ ÏÉùÏÑ±
    column_definitions = ", ".join([f"{col} {get_sqlite_type(merged_data[col])}" for col in merged_data.columns])
    cursor.execute(f"DROP TABLE IF EXISTS {MERGED_TABLE_NAME}")
    create_table_query = f"CREATE TABLE IF NOT EXISTS {MERGED_TABLE_NAME} ({column_definitions})"
    cursor.execute(create_table_query)
    conn.commit()

    # Î≥ëÌï©Îêú Îç∞Ïù¥ÌÑ∞Î•º Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§Ïóê Ï†ÄÏû•
    merged_data.to_sql(MERGED_TABLE_NAME, conn, if_exists='append', index=False)
    conn.close()
    print(f"Data saved to output database '{OUTPUT_DB}' successfully.")

# Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖÏùÑ SQLite ÌÉÄÏûÖÏúºÎ°ú Î≥ÄÌôòÌïòÎäî Ìï®Ïàò
def get_sqlite_type(series):
    if pd.api.types.is_integer_dtype(series):
        return "INTEGER"
    elif pd.api.types.is_float_dtype(series):
        return "REAL"
    elif pd.api.types.is_bool_dtype(series):
        return "BOOLEAN"
    else:
        return "TEXT"

# Î©îÏù∏ Ìï®Ïàò
if __name__ == "__main__":
    merged_data = merge_tables_from_dbs()
    save_merged_data(merged_data)
```

```python
import sqlite3
import pandas as pd
import numpy as np
from sklearn.linear_model import SGDRegressor
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
import joblib  # Î™®Îç∏ Ï†ÄÏû• Î∞è Î∂àÎü¨Ïò§Í∏∞
import re
import os
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ Î∞è Î™®Îç∏ Í≤ΩÎ°ú ÏÑ§Ï†ï
DATA_DB = "testdata.db"
MODEL_PATH = "sgd_model.pkl"

# Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è Ï†ÑÏ≤òÎ¶¨ Ìï®Ïàò Ï†ïÏùò
def load_and_preprocess_data():
    # Îç∞Ïù¥ÌÑ∞Î≤†Ïù¥Ïä§ÏóêÏÑú Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
    conn = sqlite3.connect(DATA_DB)
    data = pd.read_sql_query("SELECT * FROM merged_data", conn)
    conn.close()

    # 'sales' Ïª¨Îüº Î≥ÄÌôò
    def convert_sales(sales_value):
        if pd.isnull(sales_value) or sales_value.strip() == '':
            return 0
        try:
            sales_value = sales_value.lower().replace('k', 'e3').replace('m', 'e6')
            return int(float(re.sub(r'[^\de\.]', '', sales_value)))
        except ValueError:
            return 0

    data['sales'] = data['sales'].apply(convert_sales)

    # Ïà´ÏûêÌòï Î≥ÄÏàò Ï≤òÎ¶¨ Î∞è Í≤∞Ï∏°Ïπò Ï†úÍ±∞
    numeric_features = ['rating', 'follower', 'like', 'comments', 'bookmark', 'share', 'videocount', 'price']
    for col in numeric_features:
        data[col] = pd.to_numeric(data[col], errors='coerce')

    data = data.dropna(subset=numeric_features)

    # Îç∞Ïù¥ÌÑ∞ ÏßëÍ≥Ñ Î∞è Ï†ÑÏ≤òÎ¶¨
    grouped_data = data.groupby('name').agg({
        'rating': 'mean',
        'follower': 'mean',
        'like': 'sum',
        'comments': 'sum',
        'bookmark': 'sum',
        'share': 'sum',
        'videocount': 'sum',
        'price': 'mean',
        'hashtag': lambda x: ' '.join(x),
        'sales': 'sum'
    }).reset_index()

    # Ï†ÑÏ≤òÎ¶¨ ÌååÏù¥ÌîÑÎùºÏù∏ ÏÑ§Ï†ï
    text_features = 'hashtag'
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', StandardScaler(), numeric_features),
            ('text', TfidfVectorizer(), text_features)
        ],
        remainder='drop'
    )

    grouped_data['hashtag'] = grouped_data['hashtag'].astype(str)
    X = grouped_data.drop(columns=['name', 'sales'])
    y = grouped_data['sales']

    X_processed = preprocessor.fit_transform(X)

    return X_processed, y, preprocessor

# Î™®Îç∏ Î∂àÎü¨Ïò§Í∏∞ Î∞è Ï∂îÍ∞Ä ÌïôÏäµ
def load_model():
    if os.path.exists(MODEL_PATH):
        pipeline = joblib.load(MODEL_PATH)
        print("Î™®Îç∏ Î°úÎìú ÏôÑÎ£å.")
    else:
        # ÏÉà Î™®Îç∏Í≥º ÌååÏù¥ÌîÑÎùºÏù∏ ÏÉùÏÑ±
        model = SGDRegressor(max_iter=1000, tol=1e-3)
        _, _, preprocessor = load_and_preprocess_data()
        pipeline = Pipeline([
            ('preprocessor', preprocessor),
            ('regressor', model)
        ])
        print("ÏÉà Î™®Îç∏ ÏÉùÏÑ±.")
    return pipeline

# Ïò®ÎùºÏù∏ ÌïôÏäµ ÏàòÌñâ Ìï®Ïàò
def perform_online_learning():
    # Î™®Îç∏ Î∂àÎü¨Ïò§Í∏∞
    pipeline = load_model()

    # Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è Ï†ÑÏ≤òÎ¶¨
    X, y, _ = load_and_preprocess_data()

    # Ïò®ÎùºÏù∏ ÌïôÏäµ ÏàòÌñâ
    if isinstance(pipeline.named_steps['regressor'], SGDRegressor):
        pipeline.named_steps['regressor'].partial_fit(X, y)  # PipelineÏùò 'regressor' Îã®Í≥ÑÏóêÏÑú partial_fit Ìò∏Ï∂ú
        print("Ïò®ÎùºÏù∏ ÌïôÏäµ ÏôÑÎ£å.")
    else:
        print("partial_fitÏùÑ ÏßÄÏõêÌïòÏßÄ ÏïäÎäî Î™®Îç∏ÏûÖÎãàÎã§. Ïò®ÎùºÏù∏ ÌïôÏäµÏùÑ Í±¥ÎÑàÎúÅÎãàÎã§.")

    # Î™®Îç∏ Ï†ÄÏû•
    joblib.dump(pipeline, MODEL_PATH)
    print(f"Î™®Îç∏Ïù¥ '{MODEL_PATH}'Ïóê Ï†ÄÏû•ÎêòÏóàÏäµÎãàÎã§.")

    # ÌäπÏÑ± Ï§ëÏöîÎèÑ ÌèâÍ∞Ä ÏàòÌñâ
    evaluate_feature_importance(pipeline)

# ÌäπÏÑ± Ï§ëÏöîÎèÑ ÌèâÍ∞Ä Ìï®Ïàò
def evaluate_feature_importance(pipeline):
    if hasattr(pipeline['regressor'], 'coef_'):
        # SGDRegressorÏùò Í≤ΩÏö∞ coef_Î•º ÏÇ¨Ïö©ÌïòÏó¨ Ï§ëÏöîÎèÑ ÌèâÍ∞Ä
        coefficients = pipeline['regressor'].coef_
        feature_names = pipeline['preprocessor'].get_feature_names_out()
        importance_table = pd.DataFrame({'Feature': feature_names, 'Importance': coefficients})
        importance_table = importance_table.sort_values(by='Importance', ascending=False)

        # Ï§ëÏöîÎèÑÎ•º Í∑∏ÎûòÌîÑÎ°ú ÏãúÍ∞ÅÌôî
        plt.figure(figsize=(10, 6))
        plt.title('Feature Importance')
        plt.bar(importance_table['Feature'], importance_table['Importance'], align='center')
        plt.xticks(rotation=45)
        plt.xlabel('Feature')
        plt.ylabel('Importance')
        plt.show()

        # Ï§ëÏöîÎèÑÎ•º Ìëú ÌòïÌÉúÎ°ú Ï∂úÎ†•
        print("Feature Importance Table:")
        print(importance_table)
    else:
        print("Ìï¥Îãπ Î™®Îç∏ÏùÄ ÌäπÏÑ± Ï§ëÏöîÎèÑÎ•º ÌèâÍ∞ÄÌï† Ïàò ÏóÜÏäµÎãàÎã§.")

# Ïã§Ìñâ
if __name__ == "__main__":
    perform_online_learning()
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import SGDRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import shap
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns

# Î™®Îç∏ Î∞è Îç∞Ïù¥ÌÑ∞ Í≤ΩÎ°ú ÏÑ§Ï†ï
DATA_DB = "database_awsgd.db"

# Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è Ï†ÑÏ≤òÎ¶¨ ÌååÏù¥ÌîÑÎùºÏù∏ Ï†ïÏùò
def load_and_initialize():
    conn = sqlite3.connect(DATA_DB)
    data = pd.read_sql_query("SELECT * FROM data", conn)
    conn.close()
    return data

# Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨ ÌååÏù¥ÌîÑÎùºÏù∏ Íµ¨ÏÑ±
numeric_features = ['views', 'comments', 'likes', 'shares', 'followers']
text_features = 'hashtags'
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('text', TfidfVectorizer(), text_features)
    ],
    remainder='drop'
)

# Î™®Îç∏ Íµ¨ÏÑ± Ìï®Ïàò
def create_pipeline(model):
    return Pipeline([
        ('preprocessor', preprocessor),
        ('regressor', model)
    ])

# Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è Ï†ÑÏ≤òÎ¶¨
data = load_and_initialize()

# Ï†úÌíà-ÏòÅÏÉÅ ÏßëÍ≥Ñ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
# ÌïòÎÇòÏùò Ï†úÌíàÏóê ÎåÄÌï¥ Ïó¨Îü¨ ÏòÅÏÉÅÏù¥ Ï°¥Ïû¨Ìï† Ïàò ÏûàÏúºÎØÄÎ°ú, Ï†úÌíàÎ≥ÑÎ°ú Í¥ÄÎ†®Îêú Î™®Îì† ÏòÅÏÉÅ Îç∞Ïù¥ÌÑ∞Î•º ÏßëÍ≥Ñ
grouped_data = data.groupby('product_id').agg({
    'views': 'sum',
    'comments': 'mean',
    'likes': 'sum',
    'shares': 'sum',
    'followers': 'mean',
    'hashtags': lambda x: ' '.join(x),
    'sales_volume': 'sum'
}).reset_index()

# ÌïôÏäµÏö© Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
grouped_data['hashtags'] = grouped_data['hashtags'].astype(str)
X = grouped_data.drop(columns=['product_id', 'sales_volume'])
y = grouped_data['sales_volume']

# Î™®Îç∏ Î™©Î°ù Ï¥àÍ∏∞Ìôî
models = {
    'SGDRegressor': SGDRegressor(max_iter=1000, tol=1e-3),
    'RandomForestRegressor': RandomForestRegressor(n_estimators=100)
}

# Î™®Îç∏ ÌïôÏäµ Î∞è ÌèâÍ∞Ä
results = {}
for name, model in models.items():
    pipeline = create_pipeline(model)
    pipeline.fit(X, y)
    y_pred = pipeline.predict(X)
    mse = mean_squared_error(y, y_pred)
    r2 = r2_score(y, y_pred)
    results[name] = {'MSE': mse, 'R2': r2}
    print(f"{name} Î™®Îç∏ ÌïôÏäµ Î∞è ÌèâÍ∞Ä ÏôÑÎ£å")
    print(f"Mean Squared Error (MSE): {mse}")
    print(f"R^2 Score: {r2}")

# ÌäπÏÑ± Ï§ëÏöîÎèÑ ÌèâÍ∞Ä (1Î≤à Î™©Ìëú)
def evaluate_feature_importance(pipeline):
    if hasattr(pipeline['regressor'], 'feature_importances_'):
        importances = pipeline['regressor'].feature_importances_
        feature_names = pipeline['preprocessor'].get_feature_names_out()
        if len(feature_names) > len(importances):
            feature_names = feature_names[:len(importances)]
        indices = np.argsort(importances)[::-1]
        plt.figure(figsize=(10, 6))
        plt.title('Feature Importance')
        plt.bar(range(len(importances)), importances[indices], align='center')
        plt.xticks(range(len(importances)), [feature_names[i] for i in indices if i < len(feature_names)], rotation=45)
        plt.xlabel('Feature')
        plt.ylabel('Importance')
        plt.show()
    else:
        # SHAP Í∞íÏùÑ Ïù¥Ïö©Ìïú ÌäπÏÑ± Ï§ëÏöîÎèÑ ÌèâÍ∞Ä
        explainer = shap.Explainer(pipeline['regressor'], pipeline['preprocessor'].transform(X))
        shap_values = explainer(pipeline['preprocessor'].transform(X))
        shap.summary_plot(shap_values, X, feature_names=pipeline['preprocessor'].get_feature_names_out())

# ÎßàÏºÄÌåÖ ÏöîÏÜåÏôÄ ÌåêÎß§Îüâ Í∞Ñ ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÏãúÍ∞ÅÌôî (3Î≤à Î™©Ìëú)
def visualize_feature_correlation(data):
    numeric_data = data[numeric_features + ['sales_volume']]
    correlation_matrix = numeric_data.corr()
    plt.figure(figsize=(12, 8))
    plt.title('Correlation Matrix of Features and Sales Volume')
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
    plt.show()

# Ïò®ÎùºÏù∏ ÌïôÏäµ (4Î≤à Î™©Ìëú)
class IncrementalSGDModel:
    def __init__(self):
        self.model = SGDRegressor(max_iter=1, tol=None, warm_start=True)
        self.pipeline = create_pipeline(self.model)

    def partial_fit(self, X, y):
        sample_weight = np.ones(len(y))
        sample_weight[-int(0.2 * len(y)):] *= 1.5  # ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞Ïóê Îçî ÎÜíÏùÄ Í∞ÄÏ§ëÏπò Î∂ÄÏó¨
        self.pipeline.fit(X, y, regressor__sample_weight=sample_weight)
        print("Ïò®ÎùºÏù∏ ÌïôÏäµ ÏàòÌñâ ÏôÑÎ£å")

# Î∂ÑÏÑù Ïã§Ìñâ
optimal_pipeline = create_pipeline(models['RandomForestRegressor'])
optimal_pipeline.fit(X, y)

# 1Î≤à Î™©Ìëú: ÌäπÏÑ± Ï§ëÏöîÎèÑ ÌèâÍ∞Ä
evaluate_feature_importance(optimal_pipeline)

# 3Î≤à Î™©Ìëú: ÎßàÏºÄÌåÖ ÏöîÏÜåÏôÄ ÌåêÎß§Îüâ Í∞Ñ ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÏãúÍ∞ÅÌôî
visualize_feature_correlation(grouped_data)

# 4Î≤à Î™©Ìëú: Ïò®ÎùºÏù∏ ÌïôÏäµ ÏòàÏ†ú
incremental_model = IncrementalSGDModel()
incremental_model.partial_fit(X, y)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import SGDRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import sqlite3
import matplotlib.pyplot as plt

# Î™®Îç∏ Î∞è Îç∞Ïù¥ÌÑ∞ Í≤ΩÎ°ú ÏÑ§Ï†ï
DATA_DB = "database_awsgd.db"

# Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è Ï†ÑÏ≤òÎ¶¨ ÌååÏù¥ÌîÑÎùºÏù∏ Ï†ïÏùò
def load_and_initialize():
    conn = sqlite3.connect(DATA_DB)
    data = pd.read_sql_query("SELECT * FROM data", conn)
    conn.close()
    return data

# Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨ ÌååÏù¥ÌîÑÎùºÏù∏ Íµ¨ÏÑ±
numeric_features = ['views', 'comments', 'likes', 'shares', 'followers']
text_features = 'hashtags'
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('text', TfidfVectorizer(), text_features)
    ],
    remainder='drop'
)

# Î™®Îç∏ Íµ¨ÏÑ± Ìï®Ïàò
def create_pipeline(model):
    return Pipeline([
        ('preprocessor', preprocessor),
        ('regressor', model)
    ])

# Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è Ï†ÑÏ≤òÎ¶¨
data = load_and_initialize()

# Ï†úÌíà-ÏòÅÏÉÅ ÏßëÍ≥Ñ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
# ÌïòÎÇòÏùò Ï†úÌíàÏóê ÎåÄÌï¥ Ïó¨Îü¨ ÏòÅÏÉÅÏù¥ Ï°¥Ïû¨Ìï† Ïàò ÏûàÏúºÎØÄÎ°ú, Ï†úÌíàÎ≥ÑÎ°ú Í¥ÄÎ†®Îêú Î™®Îì† ÏòÅÏÉÅ Îç∞Ïù¥ÌÑ∞Î•º ÏßëÍ≥Ñ
grouped_data = data.groupby('product_id').agg({
    'views': 'sum',
    'comments': 'mean',
    'likes': 'sum',
    'shares': 'sum',
    'followers': 'mean',
    'hashtags': lambda x: ' '.join(x),
    'sales_volume': 'sum'
}).reset_index()

# ÌïôÏäµÏö© Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
X = grouped_data.drop(columns=['product_id', 'sales_volume'])
y = grouped_data['sales_volume']

# Î™®Îç∏ Î™©Î°ù Ï¥àÍ∏∞Ìôî
models = {
    'SGDRegressor': SGDRegressor(max_iter=1000, tol=1e-3),
    'RandomForestRegressor': RandomForestRegressor(n_estimators=100)
}

# Î™®Îç∏ ÌïôÏäµ Î∞è ÌèâÍ∞Ä
results = {}
for name, model in models.items():
    pipeline = create_pipeline(model)
    pipeline.fit(X, y)
    y_pred = pipeline.predict(X)
    mse = mean_squared_error(y, y_pred)
    r2 = r2_score(y, y_pred)
    results[name] = {'MSE': mse, 'R2': r2}
    print(f"{name} Î™®Îç∏ ÌïôÏäµ Î∞è ÌèâÍ∞Ä ÏôÑÎ£å")
    print(f"Mean Squared Error (MSE): {mse}")
    print(f"R^2 Score: {r2}")

# ÏµúÏ†Å Ïù∏ÌîåÎ£®Ïñ∏ÏÑú Í∑úÎ™® Î∂ÑÏÑù (1Î≤à Î™©Ìëú)
def analyze_optimal_influencer_size(pipeline):
    followers_range = np.linspace(500, 10000, 100)  # ÌåîÎ°úÏõå Ïàò Î≤îÏúÑ
    predicted_sales = []

    for followers in followers_range:
        example = {'views': 5000, 'comments': 50, 'likes': 200, 'shares': 20, 'followers': followers, 'hashtags': '#beauty #makeup'}
        df_example = pd.DataFrame([example])
        X_example = df_example
        sales_prediction = pipeline.predict(X_example)[0]
        predicted_sales.append(sales_prediction)

    # Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
    plt.figure(figsize=(10, 6))
    plt.plot(followers_range, predicted_sales, label='Predicted Sales')
    plt.xlabel('Number of Followers')
    plt.ylabel('Predicted Sales Volume')
    plt.title('Influencer Scale vs. Predicted Sales Volume')
    plt.legend()
    plt.grid(True)
    plt.show()

# ÌäπÏÑ± Ï§ëÏöîÎèÑ ÌèâÍ∞Ä (2Î≤à Î™©Ìëú)
def evaluate_feature_importance(pipeline):
    if hasattr(pipeline['regressor'], 'feature_importances_'):
        importances = pipeline['regressor'].feature_importances_
        feature_names = pipeline['preprocessor'].get_feature_names_out()
        if len(feature_names) > len(importances):
            feature_names = feature_names[:len(importances)]
        indices = np.argsort(importances)[::-1]
        plt.figure(figsize=(10, 6))
        plt.title('Feature Importance')
        plt.bar(range(len(importances)), importances[indices], align='center')
        plt.xticks(range(len(importances)), [feature_names[i] for i in indices if i < len(feature_names)], rotation=45)
        plt.xlabel('Feature')
        plt.ylabel('Importance')
        plt.show()
    else:
        print("Ìï¥Îãπ Î™®Îç∏ÏùÄ ÌäπÏÑ± Ï§ëÏöîÎèÑ ÌèâÍ∞Ä Í∏∞Îä•Ïù¥ ÏóÜÏäµÎãàÎã§.")

# ÎàÑÏ†Å Ìö®Í≥º Î∂ÑÏÑù (3Î≤à Î™©Ìëú)
def analyze_cumulative_effect(pipeline):
    num_videos = range(1, 21)  # ÏòÅÏÉÅ Í∞úÏàò Î≤îÏúÑ
    predicted_sales = []

    for n in num_videos:
        example = {'views': 5000 * n, 'comments': 50 * n, 'likes': 200 * n, 'shares': 20 * n, 'followers': 1000, 'hashtags': '#beauty #makeup'}
        df_example = pd.DataFrame([example])
        X_example = df_example
        sales_prediction = pipeline.predict(X_example)[0]
        predicted_sales.append(sales_prediction)

    # Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
    plt.figure(figsize=(10, 6))
    plt.plot(num_videos, predicted_sales, label='Predicted Sales with Increasing Videos')
    plt.xlabel('Number of Videos')
    plt.ylabel('Predicted Sales Volume')
    plt.title('Cumulative Effect of Multiple Videos on Sales Volume')
    plt.legend()
    plt.grid(True)
    plt.show()

# Î∂ÑÏÑù Ïã§Ìñâ
optimal_pipeline = create_pipeline(models['RandomForestRegressor'])
optimal_pipeline.fit(X, y)

# 1Î≤à Î™©Ìëú: ÏµúÏ†Å Ïù∏ÌîåÎ£®Ïñ∏ÏÑú Í∑úÎ™® Î∂ÑÏÑù
analyze_optimal_influencer_size(optimal_pipeline)

# 2Î≤à Î™©Ìëú: ÌäπÏÑ± Ï§ëÏöîÎèÑ ÌèâÍ∞Ä
evaluate_feature_importance(optimal_pipeline)

# 3Î≤à Î™©Ìëú: ÎàÑÏ†Å Ìö®Í≥º Î∂ÑÏÑù
analyze_cumulative_effect(optimal_pipeline)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import SGDRegressor
from sklearn.ensemble import RandomForestRegressor
from lightgbm import LGBMRegressor
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import sqlite3
import matplotlib.pyplot as plt

# Î™®Îç∏ Î∞è Îç∞Ïù¥ÌÑ∞ Í≤ΩÎ°ú ÏÑ§Ï†ï
DATA_DB = "database_awsgd.db"

# Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è Ï†ÑÏ≤òÎ¶¨ ÌååÏù¥ÌîÑÎùºÏù∏ Ï†ïÏùò
def load_and_initialize():
    conn = sqlite3.connect(DATA_DB)
    data = pd.read_sql_query("SELECT * FROM data", conn)
    conn.close()
    return data

# Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨ ÌååÏù¥ÌîÑÎùºÏù∏ Íµ¨ÏÑ±
numeric_features = ['views', 'comments', 'likes', 'shares', 'followers']
text_features = 'hashtags'
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('text', TfidfVectorizer(), text_features)
    ],
    remainder='drop'
)

# Î™®Îç∏ Íµ¨ÏÑ± Ìï®Ïàò
def create_pipeline(model):
    return Pipeline([
        ('preprocessor', preprocessor),
        ('regressor', model)
    ])

# Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è Ï†ÑÏ≤òÎ¶¨
data = load_and_initialize()

# Ï†úÌíà-ÏòÅÏÉÅ ÏßëÍ≥Ñ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
# ÌïòÎÇòÏùò Ï†úÌíàÏóê ÎåÄÌï¥ Ïó¨Îü¨ ÏòÅÏÉÅÏù¥ Ï°¥Ïû¨Ìï† Ïàò ÏûàÏúºÎØÄÎ°ú, Ï†úÌíàÎ≥ÑÎ°ú Í¥ÄÎ†®Îêú Î™®Îì† ÏòÅÏÉÅ Îç∞Ïù¥ÌÑ∞Î•º ÏßëÍ≥Ñ
grouped_data = data.groupby('product_id').agg({
    'views': 'sum',
    'comments': 'mean',
    'likes': 'sum',
    'shares': 'sum',
    'followers': 'mean',
    'hashtags': lambda x: ' '.join(x),
    'sales_volume': 'sum'
}).reset_index()

# ÌïôÏäµÏö© Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
X = grouped_data.drop(columns=['product_id', 'sales_volume'])
y = grouped_data['sales_volume']

# Î™®Îç∏ Î™©Î°ù Ï¥àÍ∏∞Ìôî
models = {
    'SGDRegressor': SGDRegressor(max_iter=1000, tol=1e-3),
    'RandomForestRegressor': RandomForestRegressor(n_estimators=100),
    'LGBMRegressor': LGBMRegressor(n_estimators=100)
}

# Î™®Îç∏ ÌïôÏäµ Î∞è ÌèâÍ∞Ä
results = {}
for name, model in models.items():
    pipeline = create_pipeline(model)
    pipeline.fit(X, y)
    y_pred = pipeline.predict(X)
    mse = mean_squared_error(y, y_pred)
    mae = mean_absolute_error(y, y_pred)
    r2 = r2_score(y, y_pred)
    results[name] = {'MSE': mse, 'MAE': mae, 'R2': r2}
    print(f"{name} Î™®Îç∏ ÌïôÏäµ Î∞è ÌèâÍ∞Ä ÏôÑÎ£å")
    print(f"Mean Squared Error (MSE): {mse}")
    print(f"Mean Absolute Error (MAE): {mae}")
    print(f"R^2 Score: {r2}")

# Î™®Îç∏ ÏÑ±Îä• ÎπÑÍµê Í≤∞Í≥º ÏãúÍ∞ÅÌôî
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# MSE ÎπÑÍµê
axs[0].bar(results.keys(), [results[model]['MSE'] for model in results], color='skyblue')
axs[0].set_title('Mean Squared Error (MSE) Comparison')
axs[0].set_ylabel('MSE')

# MAE ÎπÑÍµê
axs[1].bar(results.keys(), [results[model]['MAE'] for model in results], color='lightgreen')
axs[1].set_title('Mean Absolute Error (MAE) Comparison')
axs[1].set_ylabel('MAE')

# R^2 Score ÎπÑÍµê
axs[2].bar(results.keys(), [results[model]['R2'] for model in results], color='salmon')
axs[2].set_title('R^2 Score Comparison')
axs[2].set_ylabel('R^2 Score')

plt.tight_layout()
plt.show()
```
