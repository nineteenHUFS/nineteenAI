```python
import pandas as pd
import sqlite3
import re
from google.colab import files

# 텍스트 파일 경로 설정
txt_file_path = "tc.txt"

# 텍스트 파일 읽기
with open(txt_file_path, 'r', encoding='utf-8') as file:
    data = file.read()

# 정규 표현식을 사용하여 데이터 추출
pattern = re.compile(
    r'product Name: (?P<name>.*?)\n'
    r'Post \d+ Uploader: (?P<uploader>.*?)\n'
    r'Post \d+ Follower: (?P<follower>.*?)\n'
    r'Post \d+ Title: (?P<title>.*?)\n'
    r'Post \d+ URL: (?P<url>.*?)\n'
    r'Post \d+ Hashtags: (?P<hashtag>.*?)\n'
    r'Post \d+ Like: (?P<like>\d+)\n'
    r'Post \d+ Comments: (?P<comments>\d+)\n'
    r'Post \d+ Bookmark: (?P<bookmark>\d+)\n'
    r'Post \d+ Share: (?P<share>\d+)\n'
    r'Post \d+ Videocount: (?P<videocount>.*?)\n'
)

matches = pattern.finditer(data)

# 데이터를 저장할 리스트 생성
records = []

for match in matches:
    record = {
        'name': match.group('name').strip(),
        'influencer': match.group('uploader').strip(),
        'follower': match.group('follower').strip(),
        'title': match.group('title').strip(),
        'url': match.group('url').strip(),
        'hashtag': match.group('hashtag').strip(),
        'like': int(match.group('like')),
        'comments': int(match.group('comments')),
        'bookmark': int(match.group('bookmark')),
        'share': int(match.group('share')),
        'videocount': match.group('videocount').strip(),
    }
    records.append(record)

# 데이터프레임 생성
df = pd.DataFrame(records)

# 팔로워 수와 비디오 수의 단위 변환 (예: K, M 단위 변환)
def convert_to_number(value):
    value = value.upper()
    if 'K' in value:
        return float(value.replace('K', '')) * 1000
    elif 'M' in value:
        return float(value.replace('M', '')) * 1000000
    try:
        return float(value)
    except ValueError:
        return value

df['follower'] = df['follower'].apply(convert_to_number)
df['videocount'] = df['videocount'].apply(convert_to_number)

# SQLite 데이터베이스 파일로 저장
conn = sqlite3.connect("influencer_data.db")  # 데이터베이스 파일 생성 (influencer_data.db)
df.to_sql('influencer_data', conn, if_exists='replace', index=False)  # 'influencer_data' 테이블로 저장
conn.close()  # 커넥션 종료

# 데이터베이스 파일 다운로드
files.download('influencer_data.db')  # 파일 다운로드

print("텍스트 파일이 성공적으로 데이터베이스로 변환되었고, influencer_data.db 파일이 다운로드됩니다.")
```

```python
import sqlite3
import pandas as pd

# 통합할 데이터베이스 파일들의 경로를 리스트로 정의
db_files = [
    "2page_products.db",
    "3page_products.db",
    "4page_products.db",
    "5page_products.db",
    "6page_products.db",
    "7page_products.db",
    "8page_products.db",
    "9page_products.db"
]

# 통합된 데이터베이스 파일 경로 설정
OUTPUT_DB = "all_merged_database.db"  # 최종 통합된 데이터베이스 파일명

# 통합할 테이블 이름 설정
TABLE_NAME = "products"

# 출력 데이터베이스 초기화 및 테이블 생성
def initialize_output_database():
    conn = sqlite3.connect(OUTPUT_DB)
    cursor = conn.cursor()

    # 테이블 초기화
    cursor.execute(f"DROP TABLE IF EXISTS {TABLE_NAME}")

    # 기존 데이터베이스의 테이블 구조와 동일하게 새 테이블 생성
    cursor.execute(f'''
        CREATE TABLE IF NOT EXISTS {TABLE_NAME} (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            name TEXT,
            sales TEXT,
            rating REAL,
            price REAL
        )
    ''')

    conn.commit()
    conn.close()
    print(f"Output database '{OUTPUT_DB}' initialized successfully.")

# 각 데이터베이스 파일에서 데이터 읽고 통합
def merge_databases():
    all_data = []

    for db_file in db_files:
        print(f"Processing file: {db_file}")
        conn = sqlite3.connect(db_file)
        query = f"SELECT * FROM {TABLE_NAME}"
        try:
            data = pd.read_sql_query(query, conn)
            all_data.append(data)
            print(f"Successfully read data from {db_file}")
        except Exception as e:
            print(f"Failed to read from {db_file}. Error: {e}")
        finally:
            conn.close()

    if all_data:
        # 데이터 통합
        merged_data = pd.concat(all_data, ignore_index=True)
        print("Data from all databases merged successfully.")

        # 기존 데이터베이스에서 'id' 컬럼은 자동 증가되므로 제거 후 저장
        if 'id' in merged_data.columns:
            merged_data = merged_data.drop(columns=['id'])

        # 통합된 데이터를 새로운 데이터베이스에 저장
        save_to_output_database(merged_data)

# 통합된 데이터를 출력 데이터베이스에 저장
def save_to_output_database(merged_data):
    conn = sqlite3.connect(OUTPUT_DB)
    merged_data.to_sql(TABLE_NAME, conn, if_exists='append', index=False)
    conn.close()
    print(f"Data saved to output database '{OUTPUT_DB}' successfully.")

# 메인 함수
if __name__ == "__main__":
    initialize_output_database()
    merge_databases()
```

```python
import pandas as pd
import sqlite3
import re

# 텍스트 파일 경로 설정
txt_file_path = "4page_data.txt"

# 텍스트 파일 읽기
with open(txt_file_path, 'r', encoding='utf-8') as file:
    data = file.read()

# 정규 표현식을 사용하여 데이터 추출
pattern_product = re.compile(r'product Name: (.+)')
pattern_post = re.compile(
    r'Post \d+ Uploader: (?P<uploader>.+)\n'
    r'Post \d+ Follower: (?P<follower>.+)\n'
    r'Post \d+ Title: (?P<title>.+)\n'
    r'Post \d+ URL: (?P<url>.+)\n'
    r'Post \d+ Hashtags: (?P<hashtag>.+)\n'
    r'Post \d+ Like: (?P<like>\d+)\n'
    r'Post \d+ Comments: (?P<comments>\d+)\n'
    r'Post \d+ Bookmark: (?P<bookmark>\d+)\n'
    r'Post \d+ Share: (?P<share>\d+)\n'
    r'Post \d+ Videocount: (?P<videocount>.+)\n'
)

# 데이터 추출
records = []
current_product_name = None

lines = data.splitlines()
for i, line in enumerate(lines):
    product_match = pattern_product.match(line)
    if product_match:
        current_product_name = product_match.group(1).strip()
        continue

    post_match = pattern_post.match("\n".join(lines[i:i+11]))
    if post_match and current_product_name:
        record = {
            'name': current_product_name,
            'influencer': post_match.group('uploader').strip(),
            'follower': post_match.group('follower').strip(),
            'title': post_match.group('title').strip(),
            'url': post_match.group('url').strip(),
            'hashtag': post_match.group('hashtag').strip(),
            'like': int(post_match.group('like')),
            'comments': int(post_match.group('comments')),
            'bookmark': int(post_match.group('bookmark')),
            'share': int(post_match.group('share')),
            'videocount': post_match.group('videocount').strip(),
        }
        records.append(record)

# 데이터프레임 생성
df = pd.DataFrame(records)

# 데이터프레임이 비어 있는지 확인
if df.empty:
    print("데이터프레임이 비어 있습니다. 데이터 추출 과정을 확인하세요.")
else:
    # 데이터프레임 열 목록과 일부 데이터 샘플 확인
    print("추출된 데이터프레임의 열 목록:", df.columns)
    print("데이터프레임 일부 샘플:\n", df.head())

    # 팔로워 수와 비디오 수의 단위 변환 (예: K, M 단위 변환)
    def convert_to_number(value):
        value = value.upper()
        if 'K' in value:
            return float(value.replace('K', '')) * 1000
        elif 'M' in value:
            return float(value.replace('M', '')) * 1000000
        try:
            return float(value)
        except ValueError:
            return value

    # 팔로워와 비디오 수의 형식 변환 시도
    if 'follower' in df.columns:
        df['follower'] = df['follower'].apply(convert_to_number)
    if 'videocount' in df.columns:
        df['videocount'] = df['videocount'].apply(convert_to_number)

    # SQLite 데이터베이스 파일로 저장
    db_path = "tiktok_influencer_data2.db"
    conn = sqlite3.connect(db_path)  # 데이터베이스 파일 생성
    df.to_sql('influencer_data', conn, if_exists='replace', index=False)  # 'influencer_data' 테이블로 저장
    conn.close()

    print(f"데이터베이스가 '{db_path}' 경로에 성공적으로 저장되었습니다.")
```

```python
import sqlite3
import pandas as pd

# 데이터베이스 파일 경로 설정
DB_FILE_1 = "tiktok_influencer_data2.db"  # 첫 번째 데이터베이스 파일 경로
DB_FILE_2 = "all_merged_database.db"  # 두 번째 데이터베이스 파일 경로

# 테이블 이름 설정
TABLE_NAME_1 = "influencer_data"  # 첫 번째 데이터베이스에서의 테이블 이름
TABLE_NAME_2 = "products"  # 두 번째 데이터베이스에서의 테이블 이름

# 두 테이블을 불러와 'name' 컬럼을 기준으로 병합
def merge_tables_from_dbs():
    # 첫 번째 데이터베이스 연결 및 데이터 로드
    conn1 = sqlite3.connect(DB_FILE_1)
    query1 = f"SELECT * FROM {TABLE_NAME_1}"
    df1 = pd.read_sql_query(query1, conn1)
    conn1.close()
    print(f"Data from {TABLE_NAME_1} loaded successfully.")

    # 두 번째 데이터베이스 연결 및 데이터 로드
    conn2 = sqlite3.connect(DB_FILE_2)
    query2 = f"SELECT * FROM {TABLE_NAME_2}"
    df2 = pd.read_sql_query(query2, conn2)
    conn2.close()
    print(f"Data from {TABLE_NAME_2} loaded successfully.")

    # 'name' 컬럼을 기준으로 병합 (inner join)
    merged_df = pd.merge(df1, df2, on='name', how='inner')

    # 병합된 데이터 출력
    print("Merged DataFrame:")
    print(merged_df)

    return merged_df

# 병합된 데이터를 새로운 데이터베이스에 저장
OUTPUT_DB = "merged_database.db"  # 최종 통합된 데이터베이스 파일명
MERGED_TABLE_NAME = "merged_table"

def save_merged_data(merged_data):
    conn = sqlite3.connect(OUTPUT_DB)
    cursor = conn.cursor()

    # 병합된 데이터프레임을 이용하여 동적으로 CREATE TABLE 쿼리 생성
    column_definitions = ", ".join([f"{col} {get_sqlite_type(merged_data[col])}" for col in merged_data.columns])
    cursor.execute(f"DROP TABLE IF EXISTS {MERGED_TABLE_NAME}")
    create_table_query = f"CREATE TABLE IF NOT EXISTS {MERGED_TABLE_NAME} ({column_definitions})"
    cursor.execute(create_table_query)
    conn.commit()

    # 병합된 데이터를 데이터베이스에 저장
    merged_data.to_sql(MERGED_TABLE_NAME, conn, if_exists='append', index=False)
    conn.close()
    print(f"Data saved to output database '{OUTPUT_DB}' successfully.")

# 데이터 타입을 SQLite 타입으로 변환하는 함수
def get_sqlite_type(series):
    if pd.api.types.is_integer_dtype(series):
        return "INTEGER"
    elif pd.api.types.is_float_dtype(series):
        return "REAL"
    elif pd.api.types.is_bool_dtype(series):
        return "BOOLEAN"
    else:
        return "TEXT"

# 메인 함수
if __name__ == "__main__":
    merged_data = merge_tables_from_dbs()
    save_merged_data(merged_data)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import SGDRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import shap
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
import os
import re

# 모델 및 데이터 경로 설정
DATA_DB = "testdata.db"

# 데이터베이스 초기화 및 테이블 생성
def initialize_database():
    if not os.path.exists(DATA_DB):
        conn = sqlite3.connect(DATA_DB)
        cursor = conn.cursor()
        cursor.execute('''
            CREATE TABLE IF NOT EXISTS main (
                name TEXT,
                sales TEXT,
                rating REAL,
                price REAL,
                influencer TEXT,
                follower INTEGER,
                title TEXT,
                url TEXT,
                hashtag TEXT,
                like INTEGER,
                comments INTEGER,
                bookmark INTEGER,
                share INTEGER,
                videocount INTEGER
            )
        ''')
        conn.commit()
        conn.close()

initialize_database()

# 데이터 로드 및 전처리 파이프라인 정의
# 데이터 로드 및 전처리 파이프라인 정의
def load_and_initialize():
    conn = sqlite3.connect(DATA_DB, uri=True)
    data = pd.read_sql_query("SELECT * FROM merged_data", conn)
    conn.close()

    # 'sales' 컬럼에서 숫자만 추출하여 변환
    def convert_sales(sales_value):
        if pd.isnull(sales_value) or sales_value.strip() == '':
            return 0
        try:
            sales_value = sales_value.lower().replace('k', 'e3').replace('m', 'e6')
            return int(float(re.sub(r'[^\de\.]', '', sales_value)))
        except ValueError:
            return 0

    data['sales'] = data['sales'].apply(convert_sales)

    return data


# 데이터 전처리 파이프라인 구성
numeric_features = ['rating', 'follower', 'like', 'comments', 'bookmark', 'share', 'videocount', 'price']
text_features = 'hashtag'
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('text', TfidfVectorizer(), text_features)
    ],
    remainder='drop'
)

# 모델 구성 함수
def create_pipeline(model):
    return Pipeline([
        ('preprocessor', preprocessor),
        ('regressor', model)
    ])

# 데이터 로드 및 전처리
data = load_and_initialize()

# 수치 데이터 전처리: 숫자형이 아닌 데이터가 포함되어 있는 경우 처리
for col in numeric_features:
    data[col] = pd.to_numeric(data[col], errors='coerce')

data = data.dropna(subset=numeric_features)

# 제품-영상 집계 데이터 생성
# 하나의 제품에 대해 여러 영상이 존재할 수 있으므로, 제품별로 관련된 모든 영상 데이터를 집계
grouped_data = data.groupby('name').agg({
    'rating': 'mean',
    'follower': 'mean',
    'like': 'sum',
    'comments': 'sum',
    'bookmark': 'sum',
    'share': 'sum',
    'videocount': 'sum',
    'price': 'mean',
    'hashtag': lambda x: ' '.join(x),
    'sales': 'sum'
}).reset_index()

# 학습용 데이터 준비
grouped_data['hashtag'] = grouped_data['hashtag'].astype(str)
X = grouped_data.drop(columns=['name', 'sales'])
y = grouped_data['sales']

# 모델 목록 초기화
models = {
    'SGDRegressor': SGDRegressor(max_iter=1000, tol=1e-3),
    'RandomForestRegressor': RandomForestRegressor(n_estimators=100)
}

# 모델 학습 및 평가
results = {}
for name, model in models.items():
    pipeline = create_pipeline(model)
    pipeline.fit(X, y)
    y_pred = pipeline.predict(X)
    mse = mean_squared_error(y, y_pred)
    r2 = r2_score(y, y_pred)
    results[name] = {'MSE': mse, 'R2': r2}
    print(f"{name} 모델 학습 및 평가 완료")
    print(f"Mean Squared Error (MSE): {mse}")
    print(f"R^2 Score: {r2}")

# 특성 중요도 평가 (1번 목표)
def evaluate_feature_importance(pipeline):
    if hasattr(pipeline['regressor'], 'feature_importances_'):
        importances = pipeline['regressor'].feature_importances_
        feature_names = pipeline['preprocessor'].get_feature_names_out()
        if len(feature_names) > len(importances):
            feature_names = feature_names[:len(importances)]
        indices = np.argsort(importances)[::-1]
        plt.figure(figsize=(10, 6))
        plt.title('Feature Importance')
        plt.bar(range(len(importances)), importances[indices], align='center')
        plt.xticks(range(len(importances)), [feature_names[i] for i in indices if i < len(feature_names)], rotation=45)
        plt.xlabel('Feature')
        plt.ylabel('Importance')
        plt.show()
    else:
        # SHAP 값을 이용한 특성 중요도 평가
        explainer = shap.Explainer(pipeline['regressor'], pipeline['preprocessor'].transform(X))
        shap_values = explainer(pipeline['preprocessor'].transform(X))
        shap.summary_plot(shap_values, X, feature_names=pipeline['preprocessor'].get_feature_names_out())

# 마케팅 요소와 판매량 간 상관관계 시각화 (3번 목표)
def visualize_feature_correlation(data):
    numeric_data = data[numeric_features + ['sales']]
    correlation_matrix = numeric_data.corr()
    plt.figure(figsize=(12, 8))
    plt.title('Correlation Matrix of Features and Sales Volume')
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
    plt.show()

# 온라인 학습 (4번 목표)
class IncrementalSGDModel:
    def __init__(self):
        self.model = SGDRegressor(max_iter=1, tol=None, warm_start=True)
        self.pipeline = create_pipeline(self.model)

    def partial_fit(self, X, y):
        sample_weight = np.ones(len(y))
        sample_weight[-int(0.2 * len(y)):] *= 1.5  # 새로운 데이터에 더 높은 가중치 부여
        self.pipeline.fit(X, y, regressor__sample_weight=sample_weight)
        print("온라인 학습 수행 완료")

# 분석 실행
optimal_pipeline = create_pipeline(models['RandomForestRegressor'])
optimal_pipeline.fit(X, y)

# 1번 목표: 특성 중요도 평가
evaluate_feature_importance(optimal_pipeline)

# 3번 목표: 마케팅 요소와 판매량 간 상관관계 시각화
visualize_feature_correlation(grouped_data)

# 4번 목표: 온라인 학습 예제
incremental_model = IncrementalSGDModel()
incremental_model.partial_fit(X, y)
```

```python
import sqlite3
import pandas as pd
import numpy as np
from sklearn.linear_model import SGDRegressor
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
import joblib  # 모델 저장 및 불러오기
import re
import os
import matplotlib.pyplot as plt

# 데이터베이스 및 모델 경로 설정
DATA_DB = "testdata.db"
MODEL_PATH = "sgd_model.pkl"

# 데이터 로드 및 전처리 함수 정의
def load_and_preprocess_data():
    # 데이터베이스에서 데이터 불러오기
    conn = sqlite3.connect(DATA_DB)
    data = pd.read_sql_query("SELECT * FROM merged_data", conn)
    conn.close()

    # 'sales' 컬럼 변환
    def convert_sales(sales_value):
        if pd.isnull(sales_value) or sales_value.strip() == '':
            return 0
        try:
            sales_value = sales_value.lower().replace('k', 'e3').replace('m', 'e6')
            return int(float(re.sub(r'[^\de\.]', '', sales_value)))
        except ValueError:
            return 0

    data['sales'] = data['sales'].apply(convert_sales)

    # 숫자형 변수 처리 및 결측치 제거
    numeric_features = ['rating', 'follower', 'like', 'comments', 'bookmark', 'share', 'videocount', 'price']
    for col in numeric_features:
        data[col] = pd.to_numeric(data[col], errors='coerce')

    data = data.dropna(subset=numeric_features)

    # 데이터 집계 및 전처리
    grouped_data = data.groupby('name').agg({
        'rating': 'mean',
        'follower': 'mean',
        'like': 'sum',
        'comments': 'sum',
        'bookmark': 'sum',
        'share': 'sum',
        'videocount': 'sum',
        'price': 'mean',
        'hashtag': lambda x: ' '.join(x),
        'sales': 'sum'
    }).reset_index()

    # 전처리 파이프라인 설정
    text_features = 'hashtag'
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', StandardScaler(), numeric_features),
            ('text', TfidfVectorizer(), text_features)
        ],
        remainder='drop'
    )

    grouped_data['hashtag'] = grouped_data['hashtag'].astype(str)
    X = grouped_data.drop(columns=['name', 'sales'])
    y = grouped_data['sales']

    X_processed = preprocessor.fit_transform(X)

    return X_processed, y, preprocessor

# 모델 불러오기 및 추가 학습
def load_model():
    if os.path.exists(MODEL_PATH):
        pipeline = joblib.load(MODEL_PATH)
        print("모델 로드 완료.")
    else:
        # 새 모델과 파이프라인 생성
        model = SGDRegressor(max_iter=1000, tol=1e-3)
        _, _, preprocessor = load_and_preprocess_data()
        pipeline = Pipeline([
            ('preprocessor', preprocessor),
            ('regressor', model)
        ])
        print("새 모델 생성.")
    return pipeline

# 온라인 학습 수행 함수
def perform_online_learning():
    # 모델 불러오기
    pipeline = load_model()

    # 데이터 로드 및 전처리
    X, y, _ = load_and_preprocess_data()

    # 온라인 학습 수행
    if isinstance(pipeline.named_steps['regressor'], SGDRegressor):
        pipeline.named_steps['regressor'].partial_fit(X, y)  # Pipeline의 'regressor' 단계에서 partial_fit 호출
        print("온라인 학습 완료.")
    else:
        print("partial_fit을 지원하지 않는 모델입니다. 온라인 학습을 건너뜁니다.")

    # 모델 저장
    joblib.dump(pipeline, MODEL_PATH)
    print(f"모델이 '{MODEL_PATH}'에 저장되었습니다.")

    # 특성 중요도 평가 수행
    evaluate_feature_importance(pipeline)

# 특성 중요도 평가 함수
def evaluate_feature_importance(pipeline):
    if hasattr(pipeline['regressor'], 'coef_'):
        # SGDRegressor의 경우 coef_를 사용하여 중요도 평가
        coefficients = pipeline['regressor'].coef_
        feature_names = pipeline['preprocessor'].get_feature_names_out()
        importance_table = pd.DataFrame({'Feature': feature_names, 'Importance': coefficients})
        importance_table = importance_table.sort_values(by='Importance', ascending=False)

        # 중요도를 그래프로 시각화
        plt.figure(figsize=(10, 6))
        plt.title('Feature Importance')
        plt.bar(importance_table['Feature'], importance_table['Importance'], align='center')
        plt.xticks(rotation=45)
        plt.xlabel('Feature')
        plt.ylabel('Importance')
        plt.show()

        # 중요도를 표 형태로 출력
        print("Feature Importance Table:")
        print(importance_table)
    else:
        print("해당 모델은 특성 중요도를 평가할 수 없습니다.")

# 실행
if __name__ == "__main__":
    perform_online_learning()
```
